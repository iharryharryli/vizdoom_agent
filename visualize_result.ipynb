```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 6.75]

import os
import json
```

```python
result_dir = "results/ppo_record_3"
reward_history_file = os.path.join(result_dir, "reward_history")
loss_history_file = os.path.join(result_dir, "loss_history")
parameter_file = os.path.join(result_dir, "parameter.json")
```

```python
def to_float_lists(fname):
    res = []

    with open(fname) as f:
        lines = f.readlines()

        for l in lines:
            l = l.split()
            if res == []:
                for i in range(len(l)):
                    res.append([])
            for i in range(len(l)):
                res[i].append(float(l[i]))
    
    return res
```

```python
reward = to_float_lists(reward_history_file)
loss = to_float_lists(loss_history_file)
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel("# of frames trained on")
ax1.plot(reward[0], reward[2], label="average episode reward")

ax2 = ax1.twinx()
ax2.plot(loss[0], loss[2], "r", label="action loss")

fig.legend()

plt.show()
```
